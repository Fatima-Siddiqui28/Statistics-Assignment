Q5. What is sampling in statistics ?

Ans. ampling in statistics is the process of selecting a subset (sample) from a larger group (population) in order to analyze and draw conclusions about the entire population without examining every individual.

Because it's often impractical or impossible to collect data from an entire population, sampling makes research more efficient, cost-effective, and manageable.

Key Terms:
Population: The entire group you're studying (e.g., all college students in a country).

Sample: A portion of the population selected for analysis (e.g., 500 randomly chosen students).

Types of Sampling Methods:
Random Sampling
Every member of the population has an equal chance of being selected. (Most unbiased.)

Stratified Sampling
Population is divided into subgroups (strata), and random samples are taken from each group.

Systematic Sampling
Every kth member of the population is selected (e.g., every 10th person).

Cluster Sampling
Population is divided into clusters, some clusters are randomly selected, and all members of chosen clusters are surveyed.

Convenience Sampling
Samples are taken from a group that is easy to access. (Quick but often biased.)

Q6. What are the different types of sampling methods ?

Ans. There are two broad categories of sampling methods in statistics: probability sampling and non-probability sampling. Each includes several techniques, depending on the research goals and resources.

Probability Sampling â–¶
In these methods, every member of the population has a known, non-zero chance of being selected. These are more statistically reliable.

1. Simple Random Sampling
Every individual has an equal chance of being selected.
Example: Drawing names from a hat.

2. Stratified Sampling
Population is divided into subgroups (strata) based on a characteristic (e.g., age, gender), and random samples are taken from each.

Ensures representation from each subgroup.

3. Systematic Sampling-
Select every kth element from a list, starting from a randomly chosen point.

Example: Selecting every 10th person from a list of employees.

4. Cluster Sampling
Divide the population into clusters (e.g., schools, neighborhoods), randomly select some clusters, and include all members of those clusters.

Cost-effective for large populations spread over wide areas.


Non-Probability Sampling â–¶
In these methods, not all members have a known or equal chance of being selected. They're easier and quicker but may introduce bias.

1. Convenience Sampling
Choose individuals who are easiest to access.

Example: Surveying people at a mall.

2. Judgmental (Purposive) Sampling
Researcher selects participants based on their judgment about who will be most useful or representative.

Often used in qualitative research.

3. Quota Sampling
Like stratified sampling but not random. You select a specific number from each subgroup until the quota is met.

4. Snowball Sampling
Existing participants refer others to the study.

Useful for hard-to-reach or hidden populations (e.g., drug users, niche communities).

Q.17 What is the difference between variance and standard deviation ?

Ans. Variance tells you how spread out the numbers are by looking at the average of the squared differences from the average value. Because it squares the differences, the result is in squared units (like square meters or square dollars), which can make it less intuitive.

Standard deviation is the square root of the variance. It tells you, on average, how far each number is from the mean, and it uses the same units as the original dataâ€”making it easier to understand and compare.

So, variance is more of a mathematical concept, while standard deviation is what people usually use to talk about how much things vary.

Q.18 What is skewness in a dataset ?

Ans. Skewness in a dataset refers to the asymmetry in the distribution of values:

If a dataset is perfectly symmetrical, it has zero skewnessâ€”the left and right sides of the distribution mirror each other.

If it's positively skewed (right-skewed), the tail on the right side (higher values) is longer or fatter. This usually means most data points are clustered at the lower end, but there are some large values pulling the average up.

If it's negatively skewed (left-skewed), the tail on the left side (lower values) is longer. Here, most values are high, but some low outliers pull the average down.

In simple terms, skewness tells you whether the data leans more to the left or right of the average.

Q.20 Define and explain kurtosis.

It's not just about the peak heightâ€”itâ€™s more about how heavy or light the tails are (i.e., how many outliers exist).

Types of Kurtosis:

     a. Normal distribution.

         a.High peak and fat tails.
         b.More values in the center and more extreme outliers.
         c.Data is tightly clustered but with occasional large deviations.

Kurtosis helps in:

Q.21 What is the purpose of covariance ?

Ans. Covariance measures how two variables change together:

If both variables increase or decrease together, the covariance is positive.

If one increases while the other decreases, the covariance is negative.

If there's no consistent pattern, the covariance is close to zero.

Covariance helps you understand the direction of the relationship between two variablesâ€”not the strength. It's useful in statistics, finance (e.g., portfolio diversification), and data analysis to detect whether variables move together.

------------------------------------------------------

Q.22 What does correlation measure in statistics ?

Ans. Correlation measures the strength and direction of the linear relationship between two variables.
Key Points:
Positive Correlation: As one variable increases, the other also increases (and vice versa).

Negative Correlation: As one variable increases, the other decreases.

Zero Correlation: No linear relationship between the variables.

Covariance tells you if two variables move together (direction), but correlation also tells you how strongly they move together, standardized between -1 and 1.

+1 means a perfect positive linear relationship.

-1 means a perfect negative linear relationship.

0 means no linear relationship.

In short, correlation gives a clearer idea of how strongly two variables are related, whereas covariance only tells you whether they tend to move in the same direction.

------------------------------------------------------

Q.23 What is the difference between covariance and correlation ?

Ans. The main difference between covariance and correlation lies in how they measure the relationship between two variables and how they are interpreted:
Key Differences:
Interpretability: Correlation is easier to interpret because itâ€™s on a standardized scale (-1 to +1), while covariance can be hard to interpret due to its dependence on units.

Scale: Covariance is influenced by the scale of the variables, while correlation is scale-invariant.

Use Case: Correlation is more commonly used when comparing the strength and direction of relationships, while covariance is typically used in more advanced statistical modeling (like in portfolio theory or multivariate analysis).

```
#2. Write a Python program to compute the variance and standard deviation of a dataset.

import statistics

# Function to compute variance and standard deviation
def compute_stats(data):
    variance = statistics.variance(data)  # Sample variance
    std_dev = statistics.stdev(data)      # Sample standard deviation

    print(f"Dataset: {data}")
    print(f"Variance: {variance:.2f}")
    print(f"Standard Deviation: {std_dev:.2f}")

# Example dataset
data = [4, 8, 10, 15, 20]

# Call the function
compute_stats(data)
```

```
#3. Implement sampling techniques like random sampling and stratified sampling
import random
import pandas as pd

# Sample dataset
data = {
    "ID": range(1, 21),
    "Category": ["A", "B", "C", "A", "B", "C", "A", "B", "C", "A", "B", "C", "A", "B", "C", "A", "B", "C", "A", "B"],
    "Value": [random.randint(10, 100) for _ in range(20)]
}

# Convert to DataFrame
df = pd.DataFrame(data)
print("Original Dataset:\n", df)

### Random Sampling ###
def random_sampling(df, sample_size):
    return df.sample(n=sample_size, random_state=42)

# Select 5 random samples
random_sample = random_sampling(df, 5)
print("\nRandom Sample:\n", random_sample)

### Stratified Sampling ###
def stratified_sampling(df, stratify_column, sample_size_per_group):
    return df.groupby(stratify_column, group_keys=False).apply(lambda x: x.sample(sample_size_per_group, random_state=42))

# Select 2 samples per category
stratified_sample = stratified_sampling(df, "Category", 2)
print("\nStratified Sample:\n", stratified_sample)
```

```
#4. Write a Python function to calculate the range of a dataset.
def calculate_range(data):
    # Ensure data is not empty
    if len(data) == 0:
        return "Dataset is empty"

    # Calculate the range
    data_range = max(data) - min(data)
    return data_range

# Example usage:
data = [5, 8, 12, 15, 20]
range_value = calculate_range(data)
print(f"Range of the dataset: {range_value}")
```

```
#8. Write a Python script to calculate covariance between two datasets.
import numpy as np

# Example data sets
dataset_1 = [2, 4, 6, 8, 10]
dataset_2 = [1, 3, 5, 7, 9]

# Calculate covariance using numpy's cov function
cov_matrix = np.cov(dataset_1, dataset_2)

# Covariance between dataset_1 and dataset_2 is at index (0,1) or (1,0)
covariance = cov_matrix[0, 1]

print(f"Covariance between dataset_1 and dataset_2: {covariance}")
```

```
#9. Write a Python script to calculate the correlation coefficient between two datasets.
import numpy as np

# Example data sets
dataset_1 = [2, 4, 6, 8, 10]
dataset_2 = [1, 3, 5, 7, 9]

# Calculate correlation coefficient using numpy's corrcoef function
correlation_matrix = np.corrcoef(dataset_1, dataset_2)

# Correlation coefficient between dataset_1 and dataset_2 is at index (0,1) or (1,0)
correlation_coefficient = correlation_matrix[0, 1]

print(f"Correlation coefficient between dataset_1 and dataset_2: {correlation_coefficient}")
```

```
#11. Implement and compare simple random sampling and systematic sampling
import numpy as np
import matplotlib.pyplot as plt

# Create a population of 100 data points (1 to 100)
population = np.arange(1, 101)

# Set sample size
sample_size = 20

# Simple Random Sampling
simple_random_sample = np.random.choice(population, size=sample_size, replace=False)

# Systematic Sampling
k = len(population) // sample_size  # Sampling interval
start_point = np.random.randint(0, k)  # Random starting point
systematic_sample = population[start_point::k]

# Compare both samples
print(f"Simple Random Sample: {simple_random_sample}")
print(f"Systematic Sample: {systematic_sample}")

# Plot the population and the samples
plt.figure(figsize=(10, 6))

# Plot the population
plt.plot(population, np.zeros_like(population), 'k|', label='Population', markersize=10)

# Plot Simple Random Sample
plt.plot(simple_random_sample, np.zeros_like(simple_random_sample), 'bo', label='Simple Random Sample', markersize=8)

# Plot Systematic Sample
plt.plot(systematic_sample, np.zeros_like(systematic_sample), 'ro', label='Systematic Sample', markersize=8)

plt.legend()
plt.title("Comparison of Simple Random Sampling and Systematic Sampling")
plt.xlabel("Population Index")
plt.ylabel("Sample Selection")
plt.show()
```

```
#12.  Simulate data using Python and calculate its central tendency and dispersion.
import numpy as np
import pandas as pd
import scipy.stats as stats
import matplotlib.pyplot as plt

# Step 1: Simulate data using normal distribution (mean=50, std=10)
np.random.seed(42)  # For reproducibility
data = np.random.normal(loc=50, scale=10, size=1000)

# Step 2: Central Tendency
mean = np.mean(data)
median = np.median(data)
mode = stats.mode(data)[0][0]  # Mode using scipy.stats

# Step 3: Dispersion
variance = np.var(data)
std_deviation = np.std(data)

# Step 4: Display results
print(f"Mean: {mean}")
print(f"Median: {median}")
print(f"Mode: {mode}")
print(f"Variance: {variance}")
print(f"Standard Deviation: {std_deviation}")

# Step 5: Visualization
plt.figure(figsize=(10, 6))
plt.hist(data, bins=30, edgecolor='black', alpha=0.7)
plt.axvline(mean, color='r', linestyle='dashed', linewidth=2, label=f'Mean: {mean:.2f}')
plt.axvline(median, color='g', linestyle='dashed', linewidth=2, label=f'Median: {median:.2f}')
plt.axvline(mode, color='b', linestyle='dashed', linewidth=2, label=f'Mode: {mode:.2f}')
plt.legend()
plt.title("Histogram of Simulated Data with Central Tendency")
plt.xlabel("Data Values")
plt.ylabel("Frequency")
plt.show()


```

```
#15. Calculate the interquartile range (IQR) of a dataset
import numpy as np

# Simulate data (for example, 1000 data points)
np.random.seed(42)
data = np.random.normal(loc=50, scale=10, size=1000)

# Calculate the 25th percentile (Q1) and the 75th percentile (Q3)
Q1 = np.percentile(data, 25)
Q3 = np.percentile(data, 75)

# Calculate the Interquartile Range (IQR)
IQR = Q3 - Q1

# Display the results
print(f"25th Percentile (Q1): {Q1}")
print(f"75th Percentile (Q3): {Q3}")
print(f"Interquartile Range (IQR): {IQR}")
```

```
#16. Implement Z-score normalization and explain its significance
import numpy as np
import matplotlib.pyplot as plt

# Simulate data (for example, 1000 data points from a normal distribution)
np.random.seed(42)
data = np.random.normal(loc=50, scale=10, size=1000)

# Calculate the mean and standard deviation
mean = np.mean(data)
std_dev = np.std(data)

# Apply Z-score normalization
z_scores = (data - mean) / std_dev

# Display the first 10 Z-scores
print(f"First 10 Z-scores: {z_scores[:10]}")

# Visualize the original data vs. standardized data
plt.figure(figsize=(10, 6))

# Plot original data
plt.subplot(1, 2, 1)
plt.hist(data, bins=30, color='lightblue', edgecolor='black')
plt.title('Original Data')
plt.xlabel('Data Values')
plt.ylabel('Frequency')

# Plot Z-score normalized data
plt.subplot(1, 2, 2)
plt.hist(z_scores, bins=30, color='lightgreen', edgecolor='black')
plt.title('Z-score Normalized Data')
plt.xlabel('Z-score Values')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

```
#17. Compare two datasets using their standard deviations
import numpy as np
import matplotlib.pyplot as plt

# Simulate two datasets (for example, 1000 data points each)
np.random.seed(42)

# Dataset 1: Normally distributed with mean=50, std=10
data1 = np.random.normal(loc=50, scale=10, size=1000)

# Dataset 2: Normally distributed with mean=50, std=20 (wider spread)
data2 = np.random.normal(loc=50, scale=20, size=1000)

# Calculate the standard deviation of each dataset
std_dev_data1 = np.std(data1)
std_dev_data2 = np.std(data2)

# Display the standard deviations
print(f"Standard Deviation of Dataset 1: {std_dev_data1:.2f}")
print(f"Standard Deviation of Dataset 2: {std_dev_data2:.2f}")

# Visualize the datasets and their distributions
plt.figure(figsize=(10, 6))

# Plot Dataset 1
plt.subplot(1, 2, 1)
plt.hist(data1, bins=30, color='lightblue', edgecolor='black')
plt.title('Dataset 1 (Standard Deviation: {:.2f})'.format(std_dev_data1))
plt.xlabel('Data Values')
plt.ylabel('Frequency')

# Plot Dataset 2
plt.subplot(1, 2, 2)
plt.hist(data2, bins=30, color='lightgreen', edgecolor='black')
plt.title('Dataset 2 (Standard Deviation: {:.2f})'.format(std_dev_data2))
plt.xlabel('Data Values')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

```
#18. Write a Python program to visualize covariance using a heatmap
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Step 1: Create a sample dataset
np.random.seed(42)
data = {
    'Variable1': np.random.normal(loc=50, scale=10, size=1000),
    'Variable2': np.random.normal(loc=50, scale=15, size=1000),
    'Variable3': np.random.normal(loc=50, scale=5, size=1000)
}

# Create a DataFrame
df = pd.DataFrame(data)

# Step 2: Calculate the covariance matrix
cov_matrix = df.cov()

# Step 3: Plot the covariance matrix as a heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(cov_matrix, annot=True, cmap='coolwarm', fmt='.2f', cbar=True)

# Title and labels
plt.title('Covariance Matrix Heatmap', fontsize=16)
plt.xlabel('Variables', fontsize=12)
plt.ylabel('Variables', fontsize=12)

# Show the plot
plt.show()

```

```
#19. Use seaborn to create a correlation matrix for a dataset
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Step 1: Create a sample dataset
np.random.seed(42)
data = {
    'Variable1': np.random.normal(loc=50, scale=10, size=1000),
    'Variable2': np.random.normal(loc=50, scale=15, size=1000),
    'Variable3': np.random.normal(loc=50, scale=5, size=1000),
    'Variable4': np.random.normal(loc=50, scale=20, size=1000)
}

# Create a DataFrame
df = pd.DataFrame(data)

# Step 2: Calculate the correlation matrix
correlation_matrix = df.corr()

# Step 3: Visualize the correlation matrix using a heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', cbar=True, vmin=-1, vmax=1)

# Title and labels
plt.title('Correlation Matrix Heatmap', fontsize=16)
plt.xlabel('Variables', fontsize=12)
plt.ylabel('Variables', fontsize=12)

# Show the plot
plt.show()
```

```
#21. Visualize skewness and kurtosis using Python libraries like matplotlib or seaborn
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import skew, kurtosis

# Generate data
np.random.seed(42)
data = np.random.normal(loc=50, scale=10, size=1000)  # Normal distribution

# Compute skewness and kurtosis
data_skewness = skew(data)
data_kurtosis = kurtosis(data, fisher=True)  # fisher=True returns 0 for normal dist

# Plot histogram and KDE
plt.figure(figsize=(10, 6))
sns.histplot(data, kde=True, bins=30, color='skyblue', edgecolor='black')
plt.title(f'Data Distribution\nSkewness: {data_skewness:.2f}, Kurtosis: {data_kurtosis:.2f}')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()

```

```
#22. Implement the Pearson and Spearman correlation coefficients for a dataset.
import numpy as np
import pandas as pd
from scipy.stats import pearsonr, spearmanr

# Step 1: Simulate dataset
np.random.seed(42)
x = np.random.normal(loc=50, scale=10, size=100)
y = 2 * x + np.random.normal(0, 5, size=100)  # Strong linear relationship

# Step 2: Create a DataFrame
df = pd.DataFrame({'X': x, 'Y': y})

# Step 3: Compute Pearson and Spearman correlation coefficients
pearson_corr, _ = pearsonr(df['X'], df['Y'])
spearman_corr, _ = spearmanr(df['X'], df['Y'])

# Step 4: Display results
print(f"Pearson Correlation Coefficient: {pearson_corr:.4f}")
print(f"Spearman Correlation Coefficient: {spearman_corr:.4f}")
```
